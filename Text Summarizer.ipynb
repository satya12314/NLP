```python
import re
import nltk
from urllib import request
from bs4 import BeautifulSoup as bs
import heapq
```

```python
# nltk.download('punkt')
nltk.download('stopwords')
```

```python
Contents=''
url='https://en.wikipedia.org/wiki/Machine_learning'
htmlDoc=request.urlopen(url)
SoupObjec=bs(htmlDoc,'html.parser')
ParagraphContents=SoupObjec.findAll('p')
ParagraphContents
```

```python
for content in ParagraphContents:
    Contents+=content.text
print(Contents)
```

```python
cleaned_data=re.sub(r'\[[0-9]*\]',' ',Contents)
cleaned_data=re.sub(r'\s+',' ',cleaned_data)
cleaned_data
```

```python
#Creating Sentence Tokens
sent_tokens=nltk.sent_tokenize(cleaned_data)
word_token=nltk.word_tokenize(cleaned_data)
# sent_tokens
```

```python
cleaned_data=re.sub(r'[^a-zA-Z]',' ',cleaned_data)
cleaned_data=re.sub(r'\s+',' ',cleaned_data)
cleaned_data
```

```python
#frequency of words
stopwords=nltk.corpus.stopwords.words('english')
word_frequency={}
for word in word_token:
    if word not in stopwords:
        if word not in word_frequency.keys():
            word_frequency[word]=1
        else:
            word_frequency[word]+=1
```

```python
#Calculate Weighed Frequency
max_freq=max(word_frequency.values())
for word in word_frequency.keys():
    word_frequency[word]=(word_frequency[word]/max_freq)
print(word_frequency)
```

```python
#calculate sentence score
sentence_scores={}
for sentence in sent_tokens:
    for word in nltk.word_tokenize(sentence.lower()):
        if word in word_frequency.keys():
            if len(sentence.split(' '))<30:
                if sentence not in sentence_scores.keys():
                    sentence_scores[sentence]=word_frequency[word]
                else:
                    sentence_scores[sentence]+=word_frequency[word]
                    
                    
print(sentence_scores)
                    
    
```

```python
summary=heapq.nlargest(5,sentence_scores,key=sentence_scores.get)
```
